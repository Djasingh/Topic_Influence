```python
import pandas as pd
import json
```

```python
location = "../data/mgp_data/mgpnodeList_with_year_completion_updated_with_complete_dblp_link.csv"
```

```python
def json_load(file_location):
    with open(file_location, 'r') as fp:
        data = json.load(fp)
    return data
```

```python
def json_write(data, file_location):
    with open(file_location, 'w') as fp:
        json.dump(data, fp)
    return
```

```python
def read_publication_link(location):
    people=pd.read_csv(location, sep=',', lineterminator='\n', low_memory=False)
    people = people[["Id","Name","Year","University","Country","Title","MSC","BIO","MSN","Dblp_Link"]].copy()
    temp1 = pd.read_csv("../data/dblp_data/recent_link_files/mgpnodeList_with_year_completion_updated_dblpLink_227900_268000.csv", sep=',', lineterminator='\n', low_memory=False)
    temp2 = pd.read_csv("../data/dblp_data/recent_link_files/mgpnodeList_with_year_completion_updated_dblpLink_268000_268700.csv", sep=',', lineterminator='\n', low_memory=False)
    dblp_link1 = json_load("../data/dblp_data/recent_link_files/mgp_dblp_results_1.json")
    dblp_link1.pop("last_index", None)
    dblp_link1 = {int(k):v for k,v in dblp_link1.items()}
    dblp_link2 = dict(zip(temp1['Id'],temp1['dblp_link_new']))
    dblp_link3 = dict(zip(temp2['Id'],temp2['dblp_link_new']))
    dblp_link = {**dblp_link1, **dblp_link2, **dblp_link3}
    assert len(dblp_link)==people.shape[0]
    return people, dblp_link
```

```python
people, dblp_link = read_publication_link(location)
```

```python
json_write(dblp_link, "../data/dblp_data/recent_link_files/final_mgpId2link.json")
```

```python
def save_dataframe(people, dblp_link):
    people["dblp_new_link"] = people["Id"].map(dblp_link)
    people["dblp_new_link"].fillna("",inplace=True)
    people["Dblp_Link"].fillna("",inplace=True)
    people["link_same"] = people["Dblp_Link"]==people["dblp_new_link"]
    #people.to_csv("../data/mgp_data/final_mgpnodeList_with_dblp_link.csv", sep=',', index=False)
    return people
```

```python
mod_people = save_dataframe(people, dblp_link)
```
